```python
# Prints numbers down to 1
def funct1(n):
    while n > 0:
        print(n, end=" ")
        n -= 1

funct1(3)
```

```python
# Iterates through a list to find the highest value
def funct1(numbers):
    max_val = numbers[0]
    for n in numbers:
        if n > max_val:
            max_val = n
    print(max_val)

funct1([12, 45, 2, 89, 34])
```
